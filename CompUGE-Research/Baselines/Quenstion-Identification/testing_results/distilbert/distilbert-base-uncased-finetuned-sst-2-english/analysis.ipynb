```python
import pandas as pd
# open file metrics.csv
df = pd.read_csv('metrics.csv')

df['training on'] = df['training on'].map({
    'qi_webis_2022': 'webis2022',
    'qi_webis_2020': 'webis2020',
    'qi_mintaka': 'mintaka',
    'qi_beloucif': 'beloucif'
})
df['tested on'] = df['tested on'].map({
    'qi_webis_2022': 'webis2022',
    'qi_webis_2020': 'webis2020',
    'qi_mintaka': 'mintaka',
    'qi_beloucif': 'beloucif'
})

# put the third column as the first column without changing the order of the other columns
cols = df.columns.tolist()
cols = cols[2:3] + cols[:2] + cols[3:]

df = df[cols]

df.to_csv('metrics.csv', index=False)
```

```python
# sort df by f1 column
df = df.sort_values(by='f1', ascending=False)
df.to_csv('metrics.csv', index=False)
df
```
