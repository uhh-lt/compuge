```python
import pandas as pd

def dataframe_rows_shorter_than(df: pd.DataFrame, n: int) -> pd.DataFrame:
    # Filter the rows where the length of the sentence is less than n
    short_rows = df[df.apply(lambda row: len(row['sentence'].split()) < n, axis=1)]
    # Return the filtered DataFrame
    return short_rows

preds1 = pd.read_csv('sc_all_sc_webis_2022_microsoft_test_results.csv')
preds2 = pd.read_csv('sc_compsent_19_sc_webis_2022_microsoft_test_results.csv')
preds3 = pd.read_csv('sc_webis_2022_sc_webis_2022_microsoft_test_results.csv')

dataframe_rows_shorter_than(preds3, 100)
```
