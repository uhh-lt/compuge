```python
import numpy as np
import pandas as pd
import json

folder = 'webis_comparative_questions_2020_qi'
file = folder + '/english-questions.tsv'

# import webis_comparative_questions_2020_qi/english-questions.tsv
df = pd.read_csv(file, sep='\t')
df.head()
```

```python
# keep 2 columns, comp and question
df = df[['comp', 'question']]
# rename to label and text
df.columns = ['label', 'text']
df.head()
```

```python
# split df into train and test
np.random.seed(42)
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]

train.to_csv(folder + '/train.csv', sep='\t', index=False)
test.to_csv(folder + '/test.csv', sep='\t', index=False)
```
